```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv(r'C:\Users\arjun\python\ml\sonar_mine_or_rock_prediction\sonar data.csv',header=None)
sonar_data.head(5)


```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
x=sonar_data.drop(columns=60)
y=sonar_data[60]
x_train, x_test, y_train,  y_test = train_test_split(x,y,test_size=0.1, stratify=y, random_state=1)
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction,y_train)
print('accurcy:',training_data_accuracy)
```

```python
x_train_prediction = model.predict(x_test)
training_data_accuracy = accuracy_score(x_train_prediction,y_test)
print('accurcy:',training_data_accuracy)
```
